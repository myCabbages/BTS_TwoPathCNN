```python
from keras import layers
from keras.layers import Input, Conv2D, Maximum, MaxPooling2D, concatenate, Activation
from keras.models import Model
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
```

```python
def two_path(X_input, arch_type):
    if arch_type == 'input':
        #upper path
        X = Conv2D(64, (7, 7), strides=(1,1), padding='valid')(X_input)
        X1 = Conv2D(64, (7, 7), strides=(1,1), padding='valid')(X_input)
        X1 = Maximum()([X,X1])
        X1 = MaxPooling2D((4,4), strides=(1,1), padding='valid')(X1)
        
        X = Conv2D(64, (3,3), strides=(1,1), padding='valid')(X1)
        X1 = Conv2D(64, (3,3), strides=(1,1), padding='valid')(X1)
        X1 = Maximum()([X,X1])
        X1 = MaxPooling2D((2,2), strides=(1,1), padding='valid')(X1)
        
        #lower path
        X = Conv2D(160, (13, 13), strides=(1,1), padding='valid')(X_input)
        X2 = Conv2D(160, (13, 13), strides=(1,1), padding='valid')(X_input)
        X2 = Maximum()([X,X2])
        
        #concatenation
        X = concatenate([X1,X2], axis=3)
        X = Conv2D(5, (21, 21), strides=(1,1), padding='valid')(X)
        #X = Activation('softmax')(X)
    
        #model = Model(inputs=X_input, outputs=X)
        return X
    elif arch_type == 'local':
        #upper path
        X = Conv2D(64, (7, 7), strides=(1,1), padding='valid')(X_input)
        X1 = Conv2D(64, (7, 7), strides=(1,1), padding='valid')(X_input)
        X1 = Maximum()([X,X1])
        X1 = MaxPooling2D((4,4), strides=(1,1), padding='valid')(X1)
        
        #lower path
        X = Conv2D(160, (13, 13), strides=(1,1), padding='valid')(X_input)
        X2 = Conv2D(160, (13, 13), strides=(1,1), padding='valid')(X_input)
        X2 = Maximum()([X,X2])
        
        return X1, X2
    elif arch_type == 'MF':
        #upper path
        X = Conv2D(64, (7, 7), strides=(1,1), padding='valid')(X_input)
        X1 = Conv2D(64, (7, 7), strides=(1,1), padding='valid')(X_input)
        X1 = Maximum()([X,X1])
        X1 = MaxPooling2D((4,4), strides=(1,1), padding='valid')(X1)
        
        X = Conv2D(64, (3,3), strides=(1,1), padding='valid')(X1)
        X1 = Conv2D(64, (3,3), strides=(1,1), padding='valid')(X1)
        X1 = Maximum()([X,X1])
        X1 = MaxPooling2D((2,2), strides=(1,1), padding='valid')(X1)
        
        #lower path
        X = Conv2D(160, (13, 13), strides=(1,1), padding='valid')(X_input)
        X2 = Conv2D(160, (13, 13), strides=(1,1), padding='valid')(X_input)
        X2 = Maximum()([X,X2])
        #concatenation
        X = concatenate([X1,X2], axis=3)
        
        return X
```

```python
def InputCascadeCNN(shape1, shape2):
    #concatenate input and output of the 1st two path network
    X1 = Input(shape1)
    X = two_path(X1, 'input')
    
    X2 = Input(shape2)
    X = concatenate([X, X2], axis=3)
    X = two_path(X, 'input')
    X = Activation('softmax')(X)
    
    model = Model(inputs=[X1,X2], outputs=X)
    return model
```

```python
def LocalCascadeCNN(shape1, shape2):
    X1 = Input(shape1)
    X = two_path(X1, 'input')
    
    X2 = Input(shape2)
    X21,X22 = two_path(X2, 'local')
    X21 = concatenate([X,X21], axis=3)
    X = Conv2D(64, (3,3), strides=(1,1), padding='valid')(X21)
    X21 = Conv2D(64, (3,3), strides=(1,1), padding='valid')(X21)
    X21 = Maximum()([X,X21])
    X21 = MaxPooling2D((2,2), strides=(1,1), padding='valid')(X21)
    
    X22 = concatenate([X21,X22], axis=3)
    X = Conv2D(5, (21, 21), strides=(1,1), padding='valid')(X21)
    X = Activation('softmax')(X)
    model = Model(inputs=[X1,X2], outputs=X)
    return model
```

```python
ml = LocalCascadeCNN((56,56,4), (33,33,4))
ml.summary()
```

```python
def MFCascadeCNN(shape1,shape2):
    X1 = Input(shape1)
    X = two_path(X1, 'input')
    
    X2 = Input(shape2)
    X21 = two_path(X2, 'MF')
    X = concatenate([X,X21], axis=3)
    X = Conv2D(5, (21, 21), strides=(1,1), padding='valid')(X)
    X = Activation('softmax')(X)
    model = Model(inputs=[X1,X2], outputs=X)
    return model
```
